```python
from PIL import Image

# Список символов ASCII
ASCII_CHARS = ["@", "#", "S", "%", "?", "*", "+", ";", ":", ",", "."]

# меняем размер изображения
def resize_image(image, new_width=100):
    width, height = image.size
    ratio = height/width
    new_height = int(new_width * ratio)
    resized_image = image.resize((new_width, new_height))
    return(resized_image)

# переводим пиксели в чб
def grayify(image):
    grayscale_image = image.convert("L")
    return(grayscale_image)
    
# переводим чб пиксели в символы ASCII
def pixels_to_ascii(image):
    pixels = image.getdata()
    characters = "".join([ASCII_CHARS[pixel//25] for pixel in pixels])
    return(characters)    

#сама программа с проверкой правильности ввода
def main(new_width=100):
    path = input('''...........................,,....,,,,,,,..,,..,,...,,,,,,,..,,,,..,,...,............................
...........................%S...,#@@@;%S,,#?.?@@%,;@@@@+?S..+#*@;.?#,;#@S:..........................
...........................%S.::,@@@@;;@;+@:+@*+@*,+#@+:?#,:*#,@%,@*,##+@S..........................
...........................%S.#?,@@@@;.S##%.+@?;:,..%#..*#*S*#.*@S@,;@;.::..........................
...........................%S.#?,@@@@;.;@@:.,%@@#;..%#..+#%#%S.,#@*.+@:+?*,.........................
...........................%S.#?,@@@@;..#S..,:,;@%..%#..+@SS#%..?@,.;@;+#@,.........................
...........................%S.#?.@@@@;..#S..*@?+@?,*#@*:;@#?@?..?@,.,@S+#@,.........................
..........................,%S,S?:####;..S%..,?##%,:@##@+:#%+#*..*#,..+#@%S,.........................
.........................:@+;@:+@:,,,...,,....,,...,,,,..,,.,,..,,....,,.,..........................
.........................:@;;@:+@,.....Добро пожаловать в КОНВЕРТЕР КАРТИНОК В ASCII от wysiwyg!!...
.........................:@;;@:+@,.....Введите название файла:......................................
.........................:@;;@:+@,..................................................................
.........................:@;;@:+@,..................................................................
.........................:@+;@:+@,..................................................................
.........................,#;:#:+#,..................................................................
..........................,..,..,...................................................................\n''')
    try:
        image = PIL.Image.open(path)
    except:
        print(path, "не является названием файла картинки.")
        return

  
    # переводим картинку в ASCII  
    new_image_data = pixels_to_ascii(grayify(resize_image(image)))
    
    # форматируем
    pixel_count = len(new_image_data)  
    ascii_image = "\n".join([new_image_data[index:(index+new_width)] for index in range(0, pixel_count, new_width)])
    
    # печатаем результат
    print(ascii_image)
    
    # сохраняем результат как "ascii_image.txt"
    with open("ascii_image.txt", "w") as f:
        f.write(ascii_image)

main()
```
